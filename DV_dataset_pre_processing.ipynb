```python
X = df[['Height','Weight']]
```

```python
imputer = KNNImputer(n_neighbors=3, weights='uniform', metric='nan_euclidean')
```

```python
imputer.fit(X)
```

```python
Xtrans = imputer.transform(X)
```

```python
df_aux = pd.DataFrame.from_records(columns=X.columns, data = Xtrans)
```

```python
df = df_aux_1.merge(df_aux, how='outer', left_index=True, right_index=True)
```

```python
def transform(s):
    if (s['Age']>=10 and s['Age']<=22.5 ):
        return str('[10, 22.5]')
    elif (s['Age']>22.5 and s['Age']<=35 ):
        return str(']22.5,35 ]')
    elif (s['Age']>35 and s['Age']<=47.5 ):
        return str(']35,47.5]')
    elif (s['Age']>47.5 and s['Age']<=60 ):
        return str(']47.5, 60]')
    elif (s['Age']>60 and s['Age']<=72.5 ):
        return str(']60, 72.5]')
    elif (s['Age']>72.5 and s['Age']<=80 ):
        return str(']72.5, 80]')
    elif (s['Age']>80):
        return str(']80,[')
```

```python
df['Interval_Age'] = df.apply(transform, axis=1)
```
